```python
import sys
sys.path.append("./../npy")
import utilities as util
import fourier as fourier
```

```python
file_path = "../../timeseries/B101/Channel1.csv"
```

```python
#signal = util.load_time_series(file_path)
signal = util.load_time_series(file_path)[:1000]
```

```python
# Fourier Transform
freq, fft_vals = fourier.compute_fft(signal, fs=204800)
fourier.plot_spectrum(freq, fft_vals, scale='linear')       # scale='dB' or 'linear'
```

```python
# spectrogram
f, t, Sxx = fourier.compute_spectrogram(signal, fs=204800, window_size=100, overlap=50)
fourier.plot_spectrogram(f, t, Sxx, log_scale=True)
```
