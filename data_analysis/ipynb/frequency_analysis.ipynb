```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sys.path.append("./../npy")
import utilities as util
import analysis as analysis
```

```python
file_path = "../../timeseries/B101/Channel1.csv"
```

```python
df = util.load_dataframe(file_path)
df
```

```python
start_idx = 0
end_idx = 300
df_subset = df.iloc[start_idx:end_idx+1]
```

```python
serie = df_subset['singleData'].values
n = len(serie)
dt = 1  # <-- imposta il passo temporale corretto qui!

frequenze = np.fft.fftfreq(n, d=dt)
spettro = np.fft.fft(serie)

# Mostra solo le frequenze positive
maschera = frequenze > 0
# Grafico grande
plt.figure(figsize=(14, 6))  # puoi aumentare ancora se vuoi
plt.plot(frequenze[maschera], np.abs(spettro[maschera]))
plt.title('Spettro di Frequenza', fontsize=16)
plt.xlabel('Frequenza (Hz)', fontsize=14)
plt.ylabel('Ampiezza', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
