```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
sys.path.append("../npy")
import utilities as util
import fourier as fourier
```

```python
file_path = "../../timeseries/B101/Channel1.csv"
```

```python
#signal = util.load_time_series(file_path)
signal = util.load_time_series(file_path)[:1000]
```

```python
# Disegna l'autocorrelazione
fig, ax = plt.subplots(figsize=(15, 8))  # Definisce esplicitamente la figura e gli assi
lags = min(100, len(signal) // 5)  # Adatta il numero di lag automaticamente
plot_acf(signal, lags=lags, alpha=0.05, ax=ax)
plt.title('Autocorrelazione')
plt.show()
```

```python
xf_pos, yf_pos, peaks, dominant_freqs, main_freq, period = fourier.compute_dominant_frequencies(signal, sampling_rate=204800, threshold_ratio=0.2, n_peaks=10, verbose=True)
fourier.plot_frequency_spectrum(xf_pos, yf_pos, dominant_freqs)
```
