```python
os.makedirs('tb_images', exist_ok=True)


def get_text_size(font, text):
    # Restituisce (larghezza, altezza) del testo
    bbox = font.getbbox(text)
    width = bbox[2] - bbox[0]
    height = bbox[3] - bbox[1]
    return width, height

def save_table_as_image(df, filename="clean_table.jpeg", font_path=None, font_size=14, cell_padding=10):
    # Usa un font standard se non specificato
    if font_path is None:
        font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"  # Cambia su Windows/Mac

    font = ImageFont.truetype(font_path, font_size)

    # Calcola larghezza delle colonne
    col_widths = []
    for col in df.columns:
        max_len = max([get_text_size(font, str(col))[0]] +
                      [get_text_size(font, str(val))[0] for val in df[col]])
        col_widths.append(max_len + 2 * cell_padding)

    # Altezza di una riga
    row_height = get_text_size(font, "Ag")[1] + 2 * cell_padding
    table_width = sum(col_widths)
    table_height = row_height * (len(df) + 1)  # +1 per intestazione

    # Crea immagine
    image = Image.new("RGB", (table_width, table_height), "white")
    draw = ImageDraw.Draw(image)

    # Intestazioni
    x = 0
    for i, col in enumerate(df.columns):
        draw.rectangle([x, 0, x + col_widths[i], row_height], fill="#f0f0f0", outline="black")
        draw.text((x + cell_padding, cell_padding), str(col), fill="black", font=font)
        x += col_widths[i]

    # Righe
    for row_idx, row in df.iterrows():
        y = row_height * (row_idx + 1)
        x = 0
        for i, col in enumerate(df.columns):
            draw.rectangle([x, y, x + col_widths[i], y + row_height], outline="black", fill="white")
            draw.text((x + cell_padding, y + cell_padding), str(row[col]), fill="black", font=font)
            x += col_widths[i]

    # Salva immagine
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    image.save(filename)
    print(f"Tabella salvata senza margini come '{filename}'")


```
